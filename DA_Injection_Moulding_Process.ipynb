```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

#import downloaded csv as dataframe
path = 'Group2 - missingdata.csv'
df = pd.read_csv(path)
display(df)
```

```python
# using Groupby to show amounts of records for each condition including null

condition_count = df.groupby('Condition', dropna = False).size()
display(condition_count)
```

```python
# using Groupby and aggregation function to show null count for each record under each parameter 

condition_count_null = df.groupby('Condition').aggregate(lambda x: x.isnull().sum()).style.set_caption('NULL COUNT')
condition_count_null
```

```python
#To present boxplot and show outliers and how we going to remove them
colors = {'Normal1': 'blue', 'Condition4': 'green', 'Condition1': 'red'}
condition = ['Condition1', 'Condition4', 'Normal1']
df1 = df

for idx in range(12):
    sns.boxplot(x='Condition', y=df1.columns[idx+1], data=df1, hue='Condition', palette=colors, legend=False)
    plt.title(f"Boxplot of {df1.columns[idx+1]} by Condition")
    plt.show()

    #browse through each condition to see if there is outliers for specific parameter
    for i in range(len(condition)):
        # Filter the dataset for 'Normal1' condition
        df_1 = df1[df1['Condition']==condition[i]]

        # Calculate IQR 
        Q1 = df_1[df1.columns[idx+1]].quantile(0.25)
        Q3 = df_1[df1.columns[idx+1]].quantile(0.75)
        IQR = Q3 - Q1

        # Define the lower and upper bounds for outliers 
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify the outliers with tolerance of 10%
        outliers_upper = df_1[df_1[df1.columns[idx+1]] > upper_bound + upper_bound*0.1]
        outliers_lower = df_1[df_1[df1.columns[idx+1]] < lower_bound - lower_bound*0.1]
      
        # Print the number of outliers
        print(f"Upper outliers in {condition[i]}: {len(outliers_upper)}")
        print(f"Lower outliers in {condition[i]}: {len(outliers_lower)}\n")
    
        # using if else to detect if there is outliers and remove them
        if (len(outliers_upper) + len(outliers_lower)) >0:
            display(outliers_upper)&(outliers_lower)
            
            # Remove the outliers in specific condition in specific parameter
            df_clean = df_1[(df_1[df1.columns[idx+1]].isna()) | ((df_1[df1.columns[idx+1]] >= lower_bound-lower_bound*0.1) & (df_1[df1.columns[idx+1]] <= upper_bound+upper_bound*0.1))]
            
            #combine it back to the dataframe
            df1_cleaned = pd.concat([df_clean,df1[df1['Condition'] != condition[i]]]).sort_index()
            print(f"After outlier being dropped")
            
            #show boxplot after outlier has been dropped
            sns.boxplot(x='Condition', y=df1.columns[idx+1], data=df1_cleaned, hue='Condition', palette=colors, legend=False)
            plt.title(f"Boxplot of {df1.columns[idx+1]} after outliers removal")
            plt.show()
            df1=df1_cleaned
            
            #show dataframe and prove outlier has been dropped
            display(df1)
```

```python
#creating skewness, mean and median to help us see the skewness and to replace the null with either mean or median value
df_nonull = df1
parameter = ['Cylinder heating zone 1', 'Cylinder heating zone 4', 'Maximum injection pressure', 'Mould temperature control unit 1', 'Cycle time', 'Cylinder heating zone 5', 'Injection time', 'Cylinder heating zone 2', 'Dosage time', 'Cylinder heating zone 3', 'Switch-over volume', 'Material cushion']

# mean of data for each parameter
df1_mean = df1.groupby('Condition').mean().round(4).style.set_caption('MEAN')
df1_mean = df1_mean.data
display(df1_mean)

# median of data for each parameter
df1_median = df1.groupby('Condition').quantile(0.5).round(4).style.set_caption('MEDIAN')
df1_median = df1_median.data
display(df1_median)

#skewness of data for each parameter
df1_skew = df1.groupby('Condition').skew().round(4).style.set_caption('SKEW')
df1_skew = df1_skew.data
display(df1_skew)
#double for loop to run the specific skewness for parameters in the specific condition 
for i in range (3):
    for para in range (12):
        print(f"The skewness of {parameter[para]} in {condition[i]} is {df1_skew.iloc[i,para]}")
        if (df1_skew.iloc[i,para]) >= 0.5 or (df1_skew.iloc[i,para]) <= -0.5:
            print(f"Therefore we are replacing it with the median value of {df1_median.iloc[i,para]}")
            #filter only specific condition and parameter
            filtered = df1.loc[df1['Condition'] == condition[i], ['Condition', parameter[para]]]
            
            #filter the rows with null and present
            filtered_null = filtered.isnull().any(axis=1)
            filtered_null_rows = filtered[filtered_null]
            display(filtered_null_rows)
            
            # Fill null values in the filtered DataFrame with median
            filtered[parameter[para]] = filtered[parameter[para]].fillna(df1_median.iloc[i, para])

            # Now update the corresponding rows and columns in df_nonull
            df_nonull.loc[df1['Condition'] == condition[i], ['Condition', parameter[para]]] = filtered
            
        else:
            print(f"Therefore we are replacing it with the mean value of {df1_mean.iloc[i,para]}")
            #filter only specific condition and parameter
            filtered = df1.loc[df1['Condition'] == condition[i], ['Condition', parameter[para]]]
            
            #filter the rows with null and present
            filtered_null = filtered.isnull().any(axis=1)
            filtered_null_rows = filtered[filtered_null]
            display(filtered_null_rows)
            
            # Fill null values in the filtered DataFrame with mean
            filtered[parameter[para]] = filtered[parameter[para]].fillna(df1_mean.iloc[i, para])

            # Now update the corresponding rows and columns in df_nonull
            df_nonull.loc[df1['Condition'] == condition[i], ['Condition', parameter[para]]] = filtered
            
    print(f"\n")
df_new = df_nonull

display(df_new)

#save new updated file into .csv file
df_new.to_csv('Group_2_filled'+'.csv', index=False)
```

```python
# using Groupby and aggregation function to show there is no null count after filling up the null values
condition_count_null = df_new.groupby('Condition').aggregate(lambda x: x.isnull().sum()).style.set_caption('NULL COUNT')
display(condition_count_null)
```

```python
# mean of data for each parameter
df_new_mean = df_new.groupby('Condition').mean().round(4).style.set_caption('MEAN')
display(df_new_mean)

#showing change(mean) in percentage with Normal1 as reference
Condition1_change_mean = np.empty(len(parameter))
Condition4_change_mean = np.empty(len(parameter))
data_mean = df_new_mean.data
for i in range (len(parameter)):
    Condition1_change_mean[i] = round(((data_mean.iloc[0,i] - data_mean.iloc[2,i])/data_mean.iloc[2,i])*100,2)
    Condition4_change_mean[i] = round(((data_mean.iloc[1,i] - data_mean.iloc[2,i])/data_mean.iloc[2,i])*100,2)
change_mean = {'Change in %':['Condition1', 'Condition4']}
change_mean = pd.DataFrame(change_mean)
for a in range (len(parameter)):
    change_mean.insert(a+1,parameter[a],[Condition1_change_mean[a], Condition4_change_mean[a]], True) 
display(change_mean)

# median of data for each parameter
df_new_median = df_new.groupby('Condition').quantile(0.5).round(4).style.set_caption('MEDIAN')
display(df_new_median)

#showing change(median) in percentage with Normal1 as reference
Condition1_change_med = np.empty(len(parameter))
Condition4_change_med = np.empty(len(parameter))
data_med = df_new_median.data
for i in range (len(parameter)):
    Condition1_change_med[i] = round(((data_med.iloc[0,i] - data_med.iloc[2,i])/data_med.iloc[2,i])*100,2)
    Condition4_change_med[i] = round(((data_med.iloc[1,i] - data_med.iloc[2,i])/data_med.iloc[2,i])*100,2)
change_med = {'Change in %':['Condition1', 'Condition4']}
change_med = pd.DataFrame(change_med)
for a in range (len(parameter)):
    change_med.insert(a+1,parameter[a],[Condition1_change_med[a], Condition4_change_med[a]], True) 
display(change_med)

# range of data for each parameter
df_new_range = (df_new.groupby('Condition').max() - df_new.groupby('Condition').min()).style.set_caption('RANGE')
display(df_new_range)
#range of data in percentage with respect to mean
df_range_perc = (df_new_range.data / df_new_mean.data) * 100
display(df_range_perc)

# standard deviation of data for each parameter
df_new_std = df_new.groupby('Condition').std().style.set_caption('STANDARD DEVIATION')
display(df_new_std)

# min of data for each parameter
df_new_min = df_new.groupby('Condition').min().round(4).style.set_caption('MIN')
display(df_new_min)

# max of data for each parameter
df_new_max = df_new.groupby('Condition').max().round(4).style.set_caption('MAX')
display(df_new_max)
```

```python
df2 = df_new
condition = ['Normal1', 'Condition4', 'Condition1']

# Set up of 3x4 grid
fig, axes = plt.subplots(4, 3, figsize=(15, 15))  
fig.suptitle('Boxplots for Each Parameter by Condition', fontsize=16)
axes = axes.flatten()  # Flatten the 4x3 grid into a 1D array for easy indexing

# Color Definition
colors = {
    'Condition1': 'red',
    'Condition4': 'green',
    'Normal1': 'blue'

}

# Loop parameters and plot boxplot
for idx in range(12):
    param_name = df2.columns[idx+1]

    # Prepare the data for the plot
    data = [df2[df2['Condition'] == cond][param_name].dropna() for cond in condition]

    # plot with specified colors
    bplot = axes[idx].boxplot(data, patch_artist=True, labels=condition)  # Updated 'labels' to 'tick_labels'
    axes[idx].set_title(f"Boxplot of {param_name}")
    axes[idx].set_xlabel('Condition')
    axes[idx].set_ylabel(param_name)

    # Enable grid lines
    axes[idx].grid(True)

    # Apply colors to each box based on the condition
    for patch, cond in zip(bplot['boxes'], condition):
        patch.set_facecolor(colors[cond])

    # Browse through each condition to see if there are outliers for specific parameter
    for i in range(len(condition)):
        # Filter the dataset for 'Normal1', 'Condition4', 'Condition1'
        df_2 = df2[df2['Condition'] == condition[i]]
        
        # Calculate IQR
        Q1 = df_2[param_name].quantile(0.25)
        Q3 = df_2[param_name].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers with tolerance of 10%
        outliers_upper = df_2[df_2[param_name] > upper_bound + upper_bound * 0.1]
        outliers_lower = df_2[df_2[param_name] < lower_bound - lower_bound * 0.1]

        # Print the number of outliers
        print(f"Upper outliers in {condition[i]} for {param_name}: {len(outliers_upper)}")
        print(f"Lower outliers in {condition[i]} for {param_name}: {len(outliers_lower)}\n")

        # Using if else to detect if there is outliers and remove them
        if (len(outliers_upper) + len(outliers_lower)) > 0:
            print(f"Outliers detected in {condition[i]} for {param_name}, removing them...")

            # Outlier removal
            df_clean = df_2[(df_2[param_name].isna()) | ((df_2[param_name] >= lower_bound - lower_bound * 0.1) & (df_2[param_name] <= upper_bound + upper_bound * 0.1))]

            # Combine it back to the dataframe
            df2_cleaned = pd.concat([df_clean, df2[df2['Condition'] != condition[i]]]).sort_index()

            print(f"After outliers removed in {condition[i]} for {param_name}.")
            
            # Update main dataframe
            df2 = df2_cleaned


plt.tight_layout(rect=[0, 0, 1, 0.96]) 
plt.show()

```

DATA VISUALIZATION

```python
# Load CSV file
df1 = pd.read_csv('Group_2_filled.csv')
#make subplot for comparison graph 
fig, axes = plt.subplots(1, 3, figsize=(12, 6))

#make duplicate
radar1=df1.copy()
#show plot of raw data and scaled data
sns.histplot(df1['Cycle time'], kde=True, bins=15, color="blue", ax=axes[0])
axes[0].set_title("Raw data of Cycle time")
sns.histplot(df1['Cycle time']/df1['Cycle time'].max(), kde=True, bins=15, ax=axes[1])
axes[1].set_title("scalling")
# apply normalization techniques (min - max normlaisation)
for column in radar1.columns: 
    if column != 'Condition':#ensure str is not calculated
        radar1[column] = (radar1[column] - radar1[column].min()) / (radar1[column].max() - radar1[column].min())  
#show plot of normalised data
sns.histplot(radar1['Cycle time'], kde=True, bins=15, ax=axes[2],color='green')
axes[2].set_title("Normalization")
plt.tight_layout()
plt.show()
#separate out conditions
df5=radar1.query("Condition == 'Condition1'")
df6=radar1.query("Condition == 'Condition4'")
df7=radar1.query("Condition == 'Normal1'")

#put all the mean back
series1 = pd.Series(df5.iloc[:,1:].mean(), name='Condition1')
series2 = pd.Series(df6.iloc[:,1:].mean(), name='Condition4')
series3 = pd.Series(df7.iloc[:,1:].mean(), name='Normal1')

df8 = pd.concat([series1, series2, series3], axis=1)
df8 = df8.transpose()


#make 0,1,2 as index as condition is used as index
df8.index=[0,1,2]
#add back condition row as previous line deleted the line 
df8.insert(0, 'Condition',['Condition1','Condition4','Normal1'])
print("This is the mean after normalisation\n",df8)
df8=df8.loc[:,['Condition','Cylinder heating zone 1','Cylinder heating zone 2','Cylinder heating zone 3','Cylinder heating zone 4','Cylinder heating zone 5','Material cushion','Switch-over volume','Maximum injection pressure','Mould temperature control unit 1','Cycle time','Injection time','Dosage time']]
categories = list(df8.columns[1:])
N = len(categories)



# What will be the angle of each axis in the plot (we divide the plot / number of variables)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

# Initialize the radar plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Loop through each row in the dataframe and add to the radar plot
for i, row in df8.iterrows():
    values = row.drop('Condition').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=df8['Condition'])
    ax.fill(angles, values, alpha=0.1)
    
ax.legend(['Condition1','','Condition4','','Normal1'],loc='upper right',bbox_to_anchor=(1.3, 1.2))
plt.xticks(angles[:-1], categories)
#limits of the radar plot axis
ax.set_ylim(0, 0.8)
plt.show()
```

```python
# color mapping for charts
color_mapping = {
    'Normal1': 'blue',
    'Condition1': 'red',
    'Condition4': 'green'
}
num_rows = 4
num_cols = 3
```

```python
# Distribution component

# Create subplots with 4 by 3 grid
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 15))

# Flatten axes for easier iterations
axes = axes.flatten()

# List of features to plot which exclude the 'Condition' column
feature_columns = [col for col in df2.columns if col != 'Condition']

# Kernel Density Estimate plot
for i, column in enumerate(feature_columns):
    for condition, color in color_mapping.items():
        subset = df2[df2['Condition'] == condition]
        
        # Plot KDE distribution plot with the specified color
        sns.kdeplot(subset[column], color=color, ax=axes[i], label=condition, fill=True, alpha=0.3)
        
        # Calculate  median 
        median_value = subset[column].median()
        
        # Plot median line 
        axes[i].axvline(median_value, color=color, linestyle='--', label=f'{condition} Median')

    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Density')

    # Legend component
    legend = axes[i].legend(frameon=True, framealpha=0.4)  # Adjust transparency of box
    for text in legend.get_texts():
        text.set_alpha(0.7)  # Adjust text 

plt.tight_layout()
plt.show()
```

```python
# Line chart component:

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 15))

# Flatten axes for easier iterations
axes = axes.flatten()

# Create and place line plot in 3 by 4 grid
for i, column in enumerate(feature_columns):
    for condition, color in color_mapping.items():
        subset = df2[df2['Condition'] == condition]
        sns.lineplot(data=subset, x=subset.index, y=column, color=color, ax=axes[i], label=condition)
    
    axes[i].set_title(f'Line Chart of {column}')
    axes[i].set_xlabel('Index')
    axes[i].set_ylabel(column)
    axes[i].legend()

# Hide empty subplots if any
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
